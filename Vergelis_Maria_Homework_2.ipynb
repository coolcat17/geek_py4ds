**Задание 3**

```python
x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df[df['Class'] == 1].V1.head()
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
df['Class'].value_counts().plot('bar')
plt.show()
```

```python
df['Class'].value_counts().plot(kind = 'bar', logy=True)
plt.show()
```

```python
hist_1_df = df[df['Class'] == 1].V1
hist_0_df = df[df['Class'] == 0].V1
```

```python
## судя по заданию, обе гистограммы должны быть на одной оси,
## но такого примера в видео, кажется, не было,
## хотелось бы разобрать на вебинаре

fig, ax = plt.subplots(nrows=1, ncols=2)

ax1, ax2 = ax.flatten()

ax1.hist(hist_1_df, density=True, bins=20, color='red', ec='white')
ax2.hist(hist_0_df, density=True, bins=20, color='grey', ec='white')

plt.subplots_adjust(wspace=0.5)

plt.show()
```
